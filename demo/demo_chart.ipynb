```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def generate_pain_element(
    starting_pain,
    ending_pain,
    cure_time,
    total_duration,
    noise_magnitude):

    MIN_PAIN = 0
    MAX_PAIN = 10
    pain_array = []

    for i in range(cure_time):
        noise = random.randint(-noise_magnitude, noise_magnitude)
        current_pain_expected = starting_pain - (i / cure_time) * (starting_pain - ending_pain)
        current_pain_actual = round(current_pain_expected + noise)
        if current_pain_actual < MIN_PAIN:
            current_pain_actual = MIN_PAIN
        elif current_pain_actual > MAX_PAIN:
            current_pain_actual = MAX_PAIN
        pain_array.append(current_pain_actual)
        
    for i in range(cure_time, total_duration):
        noise = random.randint(-noise_magnitude, noise_magnitude)
        current_pain_expected = ending_pain
        current_pain_actual = round(current_pain_expected + noise)
        if current_pain_actual < MIN_PAIN:
            current_pain_actual = MIN_PAIN
        elif current_pain_actual > MAX_PAIN:
            current_pain_actual = MAX_PAIN
        pain_array.append(current_pain_actual)
        
    return pain_array
```

```python
def draw_demo_chart(
    starting_pain,
    ending_pain,
    cure_time,
    total_duration,
    noise_magnitude):

    x1 = np.arange(0, 60)

    y1 = generate_pain_element(
    starting_pain,
    ending_pain,
    cure_time,
    total_duration,
    noise_magnitude)

    plt.plot(x1, y1, '.-')
    plt.title('Neckare results')
    plt.ylabel('Neck pain')
    plt.xlabel('Day')

    plt.show()
```

```python
draw_demo_chart(9,0,30,60,1)
```

```python
draw_demo_chart(8,3,30,60,2)
```
